## Parte 1

Stima dell'integrale:  <font color="red">$\langle r \rangle = \int_0^1 r dr = 1/2$ </font>

Ho verificato che il valore medio di una variabile aleatoria $\,r$ distribuita uniformemente tra 0 e 1 tende a $1\over2$ all'aumentare del numero di misure di $r$.

Il grafico mostra stima e corrispondente errore (deviazione standard della media) per $\langle r \rangle$ in funzione del numero di blocchi N, con N = 100 e numero di lanci M = $10^6$. L'incertezza statistica di $\langle r\rangle$ Ã¨ calcolata utilizzato il blocking method.

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import math
import scipy.stats
import os
import subprocess

subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
subprocess.run("make", shell=True, stdout=subprocess.PIPE)
subprocess.run("make esegui", shell=True, stdout=subprocess.PIPE)

M = 10**6           # Numero di lanci
N = 100             # Numero di blocchi
L = int(M/N)        # Numero di lanci in ogni blocco (M deve essere multiplo di N)
x = np.arange(N)*L  # [0,L,2L,...,(N-1)L]

I, err = (np.genfromtxt("file1.txt", usecols=i, autostrip=True) for i in range(2))
fig = plt.figure(figsize=(14, 7) )
plt.errorbar(x,I-0.5,yerr=err)
plt.xlabel('Numero di lanci', fontsize=14)
plt.ylabel(r'$\langle r \rangle$ - 1/2', fontsize=14)
plt.grid(True)
plt.show()
```

## Parte 2

Stima dell'integrale:  <font color="red"> $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$. </font>

Il grafico mostra stima e corrispondente errore (deviazione standard della media) per $\langle \sigma^2 \rangle$ in funzione del numero di blocchi N, con N = 100 e numero di lanci M = $10^6$. Viene utilizzato lo stesso approccio della prima parte ma in questo caso viene considerata la varianza $\sigma^2 $. Sia la stima della varianza che quella del valor medio della variabile $\,r$ mostrano una buona convergenza verso il valore atteso per il numero di lanci e il numero di blocchi utilizzati.

```python
I, err = (np.genfromtxt("file2.txt", usecols=i, autostrip=True) for i in range(2))
fig = plt.figure(figsize=(14, 7) )
plt.errorbar(x,I-1/12,yerr=err)
plt.xlabel('Numero di lanci', fontsize=14)
plt.ylabel('<(r-0.5)^2>-1/12', fontsize=14)
plt.grid(True)
plt.show()
```

## Parte 3

Test del $\chi^2$ sull'intervallo $[0,1]$, diviso in $M=10^2$ sottointervalli identici. Il chi-quadro Ã¨ dato da: $$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$

Il numero di eventi attesi in ogni sotto intervallo dopo $n$ lanci per una distribuzione uniforme Ã¨ $np = n\times 1/M= n/M$. Sono stati usati per $n$ i primi $10^4$ numeri pseudo-casuali, poi i successivi $10^4$ numeri generati e cosÃ¬ via (100 volte).  
PoichÃ© in media $(n_i - n/M)^2 \simeq n/M$, allora $\chi^2 \simeq 100$ ( $\chi^2 \simeq$ numero di sotto intervalli ); un valore di $\chi^2$ grande indica che l'ipotesi che gli $n_i$ seguano una distribuzione uniforme Ã¨ improbabile. Il grafico mostra $\chi^2_j$ per $j=1, ..., 100$.

```python
# Caricamento dati da file
chi = np.genfromtxt("file3.txt", usecols=0, autostrip=True)
fig = plt.figure(figsize=(14, 8) )
plt.plot(chi, 'ro')
plt.hlines(np.mean(chi), 0, N, color = "blue", label='Mean', linestyle='--')
plt.xlabel('j', fontsize=14)
plt.ylabel(r'$\chi^2_j$', fontsize=14)
plt.grid(True)
plt.show()
```
