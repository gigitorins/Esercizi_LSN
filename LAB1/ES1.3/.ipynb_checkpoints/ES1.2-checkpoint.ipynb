```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import math
import scipy.stats

num_bins = 1000    # numero di colonne per l'istogramma
numexp = 10000
dat1u = np.zeros(numexp)
dat2e = np.zeros(numexp)
dat3l = np.zeros(numexp)

# Caricamento dati da file
file1u = open('/home/gigitorins/Scrivania/Esercizi LSN/ES1.2/file1u.txt', 'r')
file2e = open('/home/gigitorins/Scrivania/Esercizi LSN/ES1.2/file2e.txt', 'r')
file3l = open('/home/gigitorins/Scrivania/Esercizi LSN/ES1.2/file3l.txt', 'r')

for i in range(numexp):
    dat1u[i] = float(file1u.readline())
    dat2e[i] = float(file2e.readline())
    dat3l[i] = float(file3l.readline())

fig, ((ax1u, ax2e, ax3l)) = plt.subplots(nrows=1, ncols=3, figsize=(12, 6))

n, bins, columns = ax1u.hist(dat1u, num_bins)
n, bins, columns = ax2e.hist(dat2e, num_bins)
n, bins, columns = ax3l.hist(dat3l, num_bins)

# Istogramma uniforme
ax1u.set_xlabel('Bins')
ax1u.set_ylabel(r'$S_N$')
ax1u.set_title('Istogramma uniforme')

# Istogramma esponenziale
ax2e.set_xlabel('Bins')
ax2e.set_ylabel(r'$S_N$')
ax2e.set_title('Istogramma esponenziale')

# Istogramma lorentziano
ax3l.set_xlabel('Bins')
ax3l.set_ylabel(r'$S_N$')
ax3l.set_title('Istogramma lorentziano')
xlim = [-200,200]
ax3l.set_xlim(xlim)

fig.tight_layout()
plt.show()
```
