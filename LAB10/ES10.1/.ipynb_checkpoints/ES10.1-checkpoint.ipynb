```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import matplotlib.mlab as mlab
import numpy as np
import math
import scipy.stats
from itertools import cycle
from IPython.display import clear_output
import time
import scipy
from scipy.optimize import curve_fit
import subprocess
import os
import matplotlib.collections as collections
import matplotlib.ticker as ticker
```

Ho implementato il simulated annealing all'interno dell'algoritmo genetico. Viene utilizzato un algoritmo di Metropolis per arrivare, partendo da un individuo iniziale, alla soluzione del problema; l'algoritmo evolve attraverso una serie di valori di temperatura (contenuti nel parametro $\beta$) e compie un certo numero di mosse per ogni temperatura. La temperatura iniziale è relativamente alta e viene abbassata ogni volta che si compie il numero di mosse desiderato ($n_\text{moves}$). Nella simulazione $\beta$ varia in $[1, 2000]$ e $n_\text{moves}=1000$.

La probabilità di mutazione è stata impostata per tutti i tipi di mutazione ad un valore più alto rispetto all'algoritmo genetico (0.2) per compensare la mancanza del crossover nel processo di generazione di nuovi individui; questo valore si è dimostrato ottimale per poter scegliere un numero di mosse per ogni T abbastanza contenuto.

## SA per configurazione quadrata

Per le città distribuite casualmente all'interno di un quadrato, al crescere di $\beta$ (diminuzione di T) la lunghezza del best walk decresce, andando a convergere al valore minimo. I valori del best walk calcolati con l'algoritmo genetico (<a href="../../LAB9/ES9.1.ipynb">Esercitazione 9</a>) e con il simulated annealing coincidono.

```python
start = time.time()
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
arguments = "0"
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
end = time.time()
print('-- Tempo di esecuzione: {:.2f} sec '.format(end-start))
ordS = np.genfromtxt("solution.dat")
betaS, walksS = ( np.genfromtxt("walks.dat", usecols=kk, autostrip=True) for kk in range(2))
```

```python
fig = plt.figure(figsize=(10, 10))
rect = plt.Rectangle((0., 0.), 10, 10, facecolor='none', edgecolor='b',linewidth=2.0, alpha=1)
width = 0.001
h_w = 150*width
x, y = (np.genfromtxt("config_s.dat", usecols=kk, autostrip=True) for kk in range(2))

for i in range(len(x)):
    ap = int(ordS[i]-1)
    plt.plot(x[ap], y[ap],'ro')
    plt.annotate(int(ordS[i]),(x[ap],y[ap]),xytext=(x[ap]+0.01, y[ap]+0.13), size=13)
    if i == len(x)-1:
        plt.arrow(x[ap], y[ap], x[0]-x[ap], y[0]-y[ap], width=width, head_width=h_w, head_length=h_w,
                 length_includes_head=True, alpha=0.5)
    else:
        ap2 = int(ordS[i+1]-1)
        plt.arrow(x[ap], y[ap], x[ap2]-x[ap], y[ap2]-y[ap], width=width, head_width=h_w, head_length=h_w,
                  length_includes_head=True, alpha=0.5)
        
ax = plt.gca()
ax.add_patch(rect)
plt.plot(x[0],y[0],'*', markersize=16, color='orange')
plt.grid(True)  
plt.xlabel("X", fontsize=15)
plt.ylabel("Y", fontsize=15)
plt.xlim((-1,11))
plt.ylim((-1,11))
plt.title("Best walk", fontsize=17)
plt.show()
```

```python
fig = plt.figure(figsize=(15, 9))
plt.plot(betaS, walksS, label="Best walk SA")
plt.axhline(44.9525, color='r',label="Best walk AG",alpha=0.7)
print("Best walk SA: ", min(walksS))
print("Best walk AG: ", 44.9525)
plt.xlabel(r"$\beta$",fontsize=15)
plt.ylabel("Best walk",fontsize=15)
plt.legend(loc='best', prop={'size': 13})
plt.grid(True)
plt.ylim(43)
plt.show()
```

## GA per configurazione circolare

Per le città distribuite casualmente su una circonferenza, il più breve tragitto che il commesso viaggiatore può percorrere risulta sempre quello lungo la circonferenza. Al crescere di $\beta$ (diminuzione di T) la lunghezza del best walk decresce, andando a convergere al valore minimo; i valori del best walk calcolati con l'algoritmo genetico (<a href="../../LAB9/ES9.1.ipynb">Esercitazione 9</a>) e con il simulated annealing coincidono e sono inferiori alla lunghezza della circonferenza.

```python
start = time.time()
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
arguments = "1"
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
end = time.time()
print('-- Tempo di esecuzione: {:.2f} sec '.format(end-start))
ordC = np.genfromtxt("solution.dat")
betaC, walksC = ( np.genfromtxt("walks.dat", usecols=kk, autostrip=True) for kk in range(2))
```

```python
fig = plt.figure(figsize=(12, 12))
circ = plt.Circle((5., 5.), 5, facecolor='none', edgecolor='b',linewidth=2.0, alpha=1)
width = 0.001
h_w = 150*width
x, y = (np.genfromtxt("config_c.dat", usecols=kk, autostrip=True) for kk in range(2))


for i in range(len(x)):
    ap = int(ordC[i]-1)
    plt.plot(x[ap], y[ap],'ro',alpha=0.8)
    plt.annotate(int(ordC[i]),(x[ap],y[ap]),xytext=(x[ap]+0.01, y[ap]+0.13), size=13)
    if i == len(x)-1:
        plt.arrow(x[ap], y[ap], x[0]-x[ap], y[0]-y[ap], width=width, head_width=h_w, head_length=h_w,
                 length_includes_head=True, alpha=0.6)
    else:
        ap2 = int(ordC[i+1]-1)
        plt.arrow(x[ap], y[ap], x[ap2]-x[ap], y[ap2]-y[ap], width=width, head_width=h_w, head_length=h_w,
                  length_includes_head=True, alpha=0.6)
        
ax = plt.gca()
ax.add_patch(circ)
plt.plot(x[0],y[0],'*', markersize=16, color='orange')
plt.grid(True)  
plt.xlabel("X", fontsize=15)
plt.ylabel("Y", fontsize=15)
plt.xlim((-1,11))
plt.ylim((-1,11))
plt.title("Best walk", fontsize=17)
plt.show()
```

```python
fig = plt.figure(figsize=(15, 9))
plt.plot(betaC, walksC, label="Best walk SA")
plt.axhline(31.2315, color='r',label="Best walk AG",alpha=0.7)
print("Best walk SA: ", min(walksC))
print("Best walk AG: ", 31.2315)
print("Misura della circonferenza: ", 2*np.pi*5.)
plt.axhline(2*np.pi*5., color='k',label="Misura della circonferenza",alpha=0.7)
plt.xlabel(r"$\beta$",fontsize=15)
plt.ylabel("Best walk",fontsize=15)
plt.legend(loc='best', prop={'size': 13})
plt.grid(True)
plt.show()
```
