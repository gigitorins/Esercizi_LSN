# ------------------------------------------- Esercizio 11.1 -------------------------------------------

L'abilità predittiva di una rete neurale dipende fortemente dai suoi parametri, come il numero di epoche di addestramento o il numero di dati di addestramento. Utilizzando i valori forniti di default nel testo di questa esercitazione, la rete neurale performerà in modo scadente non arrivando a fittare correttamente i dati.

### Dati di input senza rumore random

Verranno utilizzati 100 dati di training e 10 di validazione (o test) con rumore nullo $\sigma=0$; le epoche di addestramento sono 30.

```python
m, b = 2, 1 # f(x) = m*x + b
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 100)
x_valid = np.random.uniform(-1, 1, 10)
x_valid.sort()
y_target = m * x_valid + b   # Funzione target

sigma = 0.0   # Dev. std. del crumore
y_train = np.random.normal(m * x_train + b, sigma)
y_valid = np.random.normal(m * x_valid + b, sigma)

fig = plt.figure(figsize =(12,6))
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True)
plt.xlabel('X', fontsize=13)
plt.ylabel('Y', fontsize=13)
plt.show()
```

## ---------------------------------------------------------- NN ----------------------------------------------------------

### Parametri iniziali

Con i parametri iniziali la loss ottenuta è molto alta anche dopo 30 epoche e la retta non viene fittata.

```python
tensorflow.keras.backend.clear_session()
model1 = models.Sequential()
model1.add(layers.Dense(1, input_shape=(1,)))
model1.summary()
model1.compile(optimizer='sgd', metrics=['mse'], loss='mse')
model1.save_weights('model1.h5')
```

```python
history = model1.fit(x=x_train, y=y_train, batch_size=32, epochs=30, shuffle=True,
                    validation_data=(x_valid, y_valid), verbose = 0)
```

```python
params = model1.get_weights()
print("Funzione: f(x) = {:03.2f} * x + {:03.2f}".format(float(params[0]), float(params[1])))
model_summary(model1)
```

# ------------------------------------------- Esercizio 11.2 -------------------------------------------

La funzione da fittare è: &nbsp;&nbsp;&nbsp; $f(x)=4-3x-2x^2+3x^3, x\in [-1,1]$

Il modello di NN deve contenere un numero ideale sia di hidden layers che di neuroni per ogni hidden layer. I dati di training generati per tale compito sono 10000; per ogni NN creato ho considerato diversi numeri per gli hidden layers $n_\text{ layers}$ e per $n_\text{ hid}$, numero di neuroni dell'hidden layer.

### Dati di input senza rumore random

```python
# Funzione target
def polynomial(x):
    return 4. - 3.*x - 2.*(x**2) + 3.*(x**3)

seed = int(time())
np.random.seed(seed)
x_train = np.random.uniform(-1, 1, 10000)
x_valid = np.random.uniform(-1, 1, 1000)
x_valid.sort()
y_target = polynomial(x_valid) 

sigma = 0.0 
y_train = np.random.normal(polynomial(x_train), sigma)
y_valid = np.random.normal(polynomial(x_valid), sigma)
```

```python
fig = plt.figure(figsize =(12,6))
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='orange')
plt.grid(True)
plt.xlabel('X', fontsize=13)
plt.ylabel('Y', fontsize=13)
plt.show()
```

## Scelta del numero di hidden layers

```python
n_lay = [1, 2, 3, 4]
vloss = []

for n in n_lay:
    tensorflow.keras.backend.clear_session()
    model2 = models.Sequential()
    model2.add(layers.Dense(4, activation='tanh'))
    for j in range(n):
        model2.add(layers.Dense(4, activation='tanh'))
    model2.add(layers.Dense(1, activation='linear'))
    model2.compile(optimizer='adam', loss='mse', metrics=['mse'])
    history = model2.fit(x=x_train, y=y_train, batch_size=32, epochs=30, shuffle=True,
          validation_data=(x_valid, y_valid), verbose = 0)
    vloss.append(history.history['loss'][-1])
```

```python
plt.subplots(figsize=(14,6))
plt.plot(n_lay, vloss)
plt.xlabel('n_hid')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```

Vediamo che utilizzare un hidden layer è la scelta migliore; l'aumento della loss per $n_\text{ layers} > 2$ indica un probabile overfitting. Infatti nella teoria delle reti neurali, il teorema di approssimazione universale afferma che una rete feed-forward con un singolo hidden layer e $n_\text{ hid}$ (numero di neuroni) finito può approssimare qualsiasi funzione continua su insiemi compatti.

Aumentare il numero di layers può portare a un ottimo grado di approssimazione della funzione target a scapito però dei costi di computazione; per la funzione che vogliamo fittare (continua) si può quindi utilizzare un singolo hidden layer. Fissiamo quindi $n_\text{ layers} = 1$ e vediamo qual è il numero di neuroni $n_\text{ hid}$ ottimale.

## Scelta del numero di neuroni per hidden layer

```python
n_hid = [5, 10, 25, 50, 100]
vloss = []

for n in n_hid:
    tensorflow.keras.backend.clear_session()
    model2 = models.Sequential()
    model2.add(layers.Dense(n, activation='tanh'))
    model2.add(layers.Dense(1, activation='linear'))
    model2.compile(optimizer='adam', loss='mse', metrics=['mse'])
    history = model2.fit(x=x_train, y=y_train, batch_size=32, epochs=30, shuffle=True,
          validation_data=(x_valid, y_valid), verbose = 0)
    vloss.append(history.history['loss'][-1])
```

```python
plt.subplots(figsize=(14,6))
plt.plot(n_hid, vloss)
plt.xlabel('n_hid')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```

In media la loss decresce al crescere del numero di neuroni dell'hidden layer, avendo però un ovvio aumento dei tempi e della complessità della rete. Si può quindi scegliere per $n_\text{hid}$ un valore intermedio tra quelli esplorati, ad esempio $n_\text{hid} = 25$.

## Scelta dell'ottimizzatore

### SGD

```python
tensorflow.keras.backend.clear_session()
model2 = models.Sequential()
model2.add(layers.Dense(25, activation='tanh'))
model2.add(layers.Dense(1, activation='linear'))

model2.compile(optimizer='sgd', metrics=['mse'], loss='mse')
history = model2.fit(x=x_train, y=y_train, batch_size=32, epochs=30, shuffle=True,
          validation_data=(x_valid, y_valid), verbose = 0)

model_summary(model2)
```

Cambiando l'ottimizzatore da SGD ad Adam, le prestazioni migliorano.

```python
tensorflow.keras.backend.clear_session()
model2 = models.Sequential()
model2.add(layers.Dense(25, activation='tanh'))
model2.add(layers.Dense(1, activation='linear'))

model2.compile(optimizer='adam', metrics=['mse'], loss='mse')
history = model2.fit(x=x_train, y=y_train, batch_size=32, epochs=30, shuffle=True,
          validation_data=(x_valid, y_valid), verbose = 0)

model_summary(model2)
```

# ------------------------------------------- Esercizio 11.3 -------------------------------------------

Fit della funzione trigonometrica $f(x,y) = \sin(x^2+y^2)$ con $x\in[-1.5, 1.5]$ e $y\in[-1.5, 1.5]$.

Per questo caso ho utilizzato una rete con 3 hidden layers da 8 neuroni e un output layer di un solo neurone con attivazione lineare; la funzione di attivazione usata per gli hidden layers è la ReLU.

```python
def sinfunc(x):
    return np.sin(x[:,0]**2. + x[:,1]**2.)

seed = int(time())
np.random.seed(seed)
ndata = 10000
nvals = int(ndata/10)

x_train = np.zeros((ndata, 2))
x_train[:,0] = np.random.uniform(-1.5, 1.5, ndata)
x_train[:,1] = np.random.uniform(-1.5, 1.5, ndata)

x_valid = np.zeros((nvals, 2))
x_valid[:,0] = np.random.uniform(-1.5, 1.5, nvals)
x_valid[:,1] = np.random.uniform(-1.5, 1.5, nvals)

y_target = sinfunc(x_valid)   # Target function

sigma = 0.0
y_train = np.zeros((ndata,1))
y_train[:,0] = np.random.normal(sinfunc(x_train), sigma)
y_valid = np.random.normal(sinfunc(x_valid), sigma)
```

```python
fig = plt.figure(figsize=(8,8))
ax = Axes3D(fig)
ax.scatter(x_valid[:,0], x_valid[:,1], y_target, marker='o', c=y_target)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(30, 60)
plt.show()
```

## ---------------------------------------------------------- NN ----------------------------------------------------------

```python
tensorflow.keras.backend.clear_session()
model3 = models.Sequential()
model3.add(layers.Dense(8, input_shape=(2,), activation='relu'))
model3.add(layers.Dense(8, activation='relu'))
model3.add(layers.Dense(8, activation='relu'))
model3.add(layers.Dense(1,activation='linear'))

model3.compile(optimizer='adam', loss='mse', metrics=['mse'])
Es = EarlyStopping(monitor='val_loss', patience=10, verbose=True)
```

```python
history = model3.fit(x = x_train, y = y_train, batch_size = 32, epochs = 100, shuffle = True,
                    validation_data = (x_valid, y_valid), verbose = 0, callbacks = [Es])
```

```python
# Plot training & validation loss values
fig = plt.figure(figsize =(15,8))
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.grid(True)
plt.title('Model loss', fontsize=16)
plt.xlabel('Epoch', fontsize=15)
plt.ylabel('Loss', fontsize=15)
plt.legend(loc=0, prop={'size': 14})
plt.show()
```
