- campionamento con importance sampling in $[0,1]$

## Distribuzione uniforme vs Importance sampling

Si può considerare l'integrale nella forma: &nbsp;&nbsp;$ I = \int_0^1 g(x)p(x) dx $&nbsp; dove $g(x) = \frac{\pi}{2}\cos(\frac{\pi}{2}x)$ e $p(x)$ è una distribuzione di probabilità.

L'integrale si può calcolare valutando il valore di $g(x)p(x)$ in $L$ punti (per ognuno degli $N$ blocchi) estratti da $p(x)$ e considerandone poi la media:
$$I = \frac{1}{L}\sum_{i=1}^{L}g(x_i)p(x_i)$$

Il blocking method fornisce una stima per $I$ e la sua incertezza al crescere del numero di blocchi considerati. Utilizzando come $p(x)$ una PDF uniforme tra 0 e 1 si ha una buona convergenza al valore atteso.

Per quanto riguarda l'importance sampling (campionamento rilevante), si deve ricercare una PDF più adatta per stimare lo stesso integrale $I$ che minimizzi la varianza $\sigma^2_I$ e garantisca una convergenza più veloce. L'integrale può essere riscritto come:
$$ I=\int_0^1 \frac{g(x)p(x)}{d(x)} d(x)dx $$
in cui $d(x)$ è una PDF e $\frac{g(x)p(x)}{d(x)}<\infty$ eccetto che su un insieme di punti di misura zero.

Utilizzando il blocking method si ha una stima di $I$:
$$G_L = \frac{1}{L}\sum_{i=1}^L\frac{g(x_i)p(x_i)}{d(x_i)}$$

Ho campionato usando $d(x) = 2(1-x)$ come distribuzione; con il metodo della trasformata si ottiene la variabile aleatoria $x = 1 - \sqrt{(1-y)}$ con y uniforme tra 0 e 1. Si nota come la stima di $I$ migliori sensibilmente rispetto al campionamento con una PDF uniforme.

```python
fig = plt.figure(figsize=(15, 8))
I, err = (np.genfromtxt("file1.txt", usecols=i, autostrip=True) for i in range(2))
plt.errorbar(x,I-1,yerr=err, label='Distribuzione uniforme')
I, err = (np.genfromtxt("file2.txt", usecols=i, autostrip=True) for i in range(2))
plt.errorbar(x,I-1,yerr=err, label='Importance sampling')
plt.axhline(0, color='red', alpha=0.6)
plt.xlabel('N° lanci', fontsize=14)
plt.ylabel('<I>-1', fontsize=14)
plt.grid(True)
plt.legend(loc=0)
plt.show()
```
