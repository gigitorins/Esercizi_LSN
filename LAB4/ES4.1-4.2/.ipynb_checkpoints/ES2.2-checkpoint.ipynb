```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import math
import scipy
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset

N = 101
x = np.arange(N)

d = np.zeros(N)
c = np.zeros(N)
med = np.zeros(N)
errd = np.zeros(N)
errc = np.zeros(N)
k = 0.92

# Caricamento dati da file
file1 = open('/home/gigitorins/Scrivania/Esercizi LSN/ES2.2/rw3dpD.txt', 'r')
file2 = open('/home/gigitorins/Scrivania/Esercizi LSN/ES2.2/rw3dpC.txt', 'r')

for i in range(N):
    d[i], errd[i] = file1.readline().split()
    c[i], errc[i] = file2.readline().split()
    med[i] = (d[i]+c[i])/2;

def test_func(x, a, b):
    return a * x**(b)

params, params_covariance = scipy.optimize.curve_fit(test_func, x, d,
                                               p0=[0.92, 0.5])

print(params)

fig, ax = plt.subplots(figsize=[15,7])

plt.xlabel('N° passi')
plt.ylabel(r'$\sqrt{<|\vec{r}_N|^2_{RW}>}$')

# sub region of the original image
x1, x2, y1, y2 = 93.0, 100.5, 8.7, 9.4
axins = zoomed_inset_axes(ax, 7.3, loc=4) # zoom = 2

axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)
plt.xticks(visible=False)
plt.yticks(visible=False)
mark_inset(ax, axins, loc1=2, loc2=1, fc="none", ec="0.5")
ax.grid(True)
ax.errorbar(x,d,yerr=errd,label='RW discreto')
ax.errorbar(x,c,yerr=errc,label='RW continuo')
ax.plot(x, test_func(x, params[0], params[1]),
         label='Previsione teorica')
plt.errorbar(x,d,yerr=errd,label='RW discreto')
plt.errorbar(x,c,yerr=errc,label='RW continuo')
plt.plot(x, test_func(x, params[0], params[1]),
         label='Previsione teorica')
plt.grid(True)
plt.legend(loc=2)
plt.draw()
plt.show()
```
