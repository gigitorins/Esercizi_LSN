```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import math
import scipy.stats
from itertools import cycle
from mpl_toolkits.mplot3d import Axes3D
import scipy
import os

currentDirectory = os.getcwd()
filename = "%s/COPIA/config.0" % currentDirectory
filename1 = "%s/COPIA/config.final" % currentDirectory

# Caricamento dati da file
x0, y0, z0 = (np.genfromtxt(filename, usecols=i, autostrip=True) for i in range(3))
x, y, z = (np.genfromtxt(filename1, usecols=i, autostrip=True) for i in range(3))

# Grafici 3D
fig = plt.figure(figsize=(12, 6) , dpi=150)

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.view_init(15, 30)
ax1.grid(True)
ax1.xaxis.pane.fill = ax1.yaxis.pane.fill = ax1.zaxis.pane.fill = False
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.scatter3D(x0, y0, z0, c='blue', alpha=0.7,s=3)
ax1.set_title('Configurazione iniziale', fontsize=13, y=0.)

ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.view_init(15, 30)
ax2.grid(True)
ax2.xaxis.pane.fill = ax2.yaxis.pane.fill = ax2.zaxis.pane.fill = False
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.scatter3D(x, y, z, c='red', alpha=0.7,s=3)
ax2.set_title('Configurazione finale', fontsize=13, y=0.)

fig.suptitle('Configurazioni 3D', fontsize=15)
fig.tight_layout()
plt.draw()
plt.show()
```

```python
npart = np.arange(108)

fig = plt.figure(figsize=(11, 7) , dpi=130)
plt.xlabel('N. particella')
plt.ylabel('DIfferenza')
plt.plot(npart, x-x0, 'k', label='X')
plt.plot(npart, y-y0, 'r', label='Y')
plt.plot(npart, z-z0, 'c', label='Z')
plt.grid(True)
plt.legend(loc=1)
plt.draw()
plt.show()
```

```python
M=10**3               # Numero di time steps
x = np.arange(M)      # [0,L,2L,...,(N-1)L]

# Caricamento dati da file
filename2 = "%s/COPIA/output_etot.dat" % currentDirectory
filename3 = "%s/COPIA/output_epot.dat" % currentDirectory
filename4 = "%s/COPIA/output_ekin.dat" % currentDirectory
filename5 = "%s/COPIA/output_temp.dat" % currentDirectory
etot = (np.genfromtxt(filename2, usecols=0, autostrip=True))
epot = (np.genfromtxt(filename3, usecols=0, autostrip=True))
ekin = (np.genfromtxt(filename4, usecols=0, autostrip=True))
temp = (np.genfromtxt(filename5, usecols=0, autostrip=True))

fig = plt.figure(figsize=(11, 7) , dpi=130)
plt.xlabel('Time step')
plt.ylabel('Energia')
plt.plot(x, etot, 'k', label='Energia totale')
plt.plot(x, epot, 'r', label='Energia potenziale')
plt.plot(x, ekin, 'c', label='Energia cinetica')
plt.plot(x, temp, 'y', label='Temperatura')
plt.grid(True)
plt.legend(loc=1)
fig.tight_layout()
plt.draw()
plt.show()
```
