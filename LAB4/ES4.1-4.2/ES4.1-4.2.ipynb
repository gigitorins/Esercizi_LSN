Utilizzando il codice di dinamica molecolare con un modello di Lennard-Jones, ho implementato la possibilità di ricominciare la simulazione di dinamica molecolare leggendo da file non solo la configurazione $\vec{r}(t)$ al tempo $t$ ma anche la configurazione $\vec r(t-dt)$ allo step precedente. Con tali configurazioni si può calcolare $\vec r(t+dt)$.

Il restart della simulazione è implementato da una variabile booleana inserita nel file <font face="Courier" size=3>input.dat</font>, che può valere $1$ o $0$. Il valore $1$ fa ripartire da una simulazione precedente, il valore $0$ fa sì che la simulazione riparta dal tempo iniziale (e quindi legga solo il file <font face="Courier" size=3>config.0</font>).

Ho inoltre inserito la possibilità di riscalare le velocità in modo che il sistema raggiunga la temperatura desiderata. Il fattore di riscalamento ($fs$) della velocità si ottiene calcolando la temperatura

$$T(t+dt/2)=\frac{\sum_{i=1}^N \mid\vec{v}_i(t+dt/2)\mid^2}{3N}$$

dove $N$ è il numero di particelle la cui dinamica viene considerata e $\vec{v}_i$ è la velocità della singola particella. La velocità al tempo $t+dt/2$ viene calcolata utilizzando gli sviluppi di Taylor di $\vec r(t+dt)$ e $\vec r(t)$ intorno a $t+dt/2$ e invertendoli:

$$\vec v(t+dt/2) =\frac{\vec r(t+dt)-\vec r(t)}{dt}$$


Il fattore di riscalamento vale quindi:

$$fs=\sqrt\frac{3T^*}{3T(t+dt/2)} = \sqrt\frac{T^*}{T(t+dt/2)}$$

dove $T^*$ è la temperatura desiderata (presente nel file <font face="Courier" size=3>input.dat</font>).

**In tutti i grafici sono riportate quantità espresse in unità di misura di Lennard-Jones**.

```python
gnames = ['Energia potenziale','Energia cinetica','Energia totale','Temperatura']
fnames = ['output_epot.dat', 'output_ekin.dat', 'output_etot.dat', 'output_temp.dat']
fig = plt.figure(figsize=(16, 10))

for j in range(4):
    y = np.genfromtxt(fnames[j], usecols=0, autostrip=True)
    x = np.arange(0,len(y))*10
    plt.plot(x, y, label=gnames[j])
    
plt.xlabel('Time step', fontsize=14)
plt.ylabel('L-J units', fontsize=14)
plt.grid(True)
plt.legend(loc=0)
plt.show()
```

## Valori medi delle grandezze termodinamiche

Ho implementato il data blocking nel codice di dinamica molecolare. I $10^4$ time-steps sono stati divisi in $N = 100$ blocchi per calcolare la media e l'incertezza statistica delle grandezze di interesse al crescere di $N$.

Il grafico sottostante mostra l'andamento di quattro grandezze termodinamiche durante la quinta e ultima simulazione; i loro valori medi si stabilizzano attorno ai valori istantanei successivi a 400 steps presenti nel grafico in alto. La temperatura converge a un valore molto simile a quello di imput e l'$E_{tot}$ rimane pressoché costante anche per quanto riguarda il suo valore medio.

```python
gnames = ['Energia potenziale','Energia cinetica','Energia totale','Temperatura']
fnames = ['ave_epot.out', 'ave_ekin.out', 'ave_etot.out', 'ave_temp.out']
fig = plt.figure(figsize=(16, 32))
x = np.arange(0,N)

for i in range(4):
    ax = fig.add_subplot(4, 1, i+1)
    y, yerror = (np.genfromtxt(fnames[i], usecols=j, autostrip=True) for j in range(2))
    ax.errorbar(x, y[-N:], yerr=yerror[-N:])
    ax.set_xlabel('Numero di blocchi',fontsize=14)
    ax.set_ylabel('L-J units',fontsize=14)
    ax.set_title(gnames[i], fontsize=16)
    ax.grid(True)

fig.tight_layout()
plt.show()
```
