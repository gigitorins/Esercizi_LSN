### Molecular Dynamics

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import math
import scipy.stats
from itertools import cycle
from mpl_toolkits.mplot3d import Axes3D
import scipy
import os

currentDirectory = os.getcwd()
filename1 = "%s/nsim.dat" % currentDirectory
nsim = np.genfromtxt(filename1, usecols=0) - 1
M = 10**4                # Numero di time-steps
N = 200                  # Numero di blocchi
x = np.arange(0,nsim*M,step=10)

# Caricamento dati da file
filename2 = "%s/output_etot.dat" % currentDirectory
filename3 = "%s/output_epot.dat" % currentDirectory
filename4 = "%s/output_ekin.dat" % currentDirectory
filename5 = "%s/output_temp.dat" % currentDirectory
etot = np.genfromtxt(filename2, usecols=0, autostrip=True)
epot = np.genfromtxt(filename3, usecols=0, autostrip=True)
ekin = np.genfromtxt(filename4, usecols=0, autostrip=True)
temp = np.genfromtxt(filename5, usecols=0, autostrip=True)

fig = plt.figure(figsize=(11, 7) , dpi=130)
plt.xlabel('Time step')
plt.plot(x, etot, 'k', label='Energia totale')
plt.plot(x, epot, 'r', label='Energia potenziale')
plt.plot(x, ekin, 'c', label='Energia cinetica')
plt.plot(x, temp, 'y', label='Temperatura')
plt.grid(True)
plt.legend(loc=0)
fig.tight_layout()
plt.draw()
plt.show()
```

```python
x = np.arange(nsim*N)

# Caricamento dati da file
filename6 = "%s/ave_epot.out" % currentDirectory
filename7 = "%s/ave_ekin.out" % currentDirectory
filename8 = "%s/ave_etot.out" % currentDirectory
filename9 = "%s/ave_temp.out" % currentDirectory
epot, errepot = (np.genfromtxt(filename6, usecols=i, autostrip=True) for i in range(2))
ekin, errekin = (np.genfromtxt(filename7, usecols=i, autostrip=True) for i in range(2))    
etot, erretot = (np.genfromtxt(filename8, usecols=i, autostrip=True) for i in range(2))
temp, errtemp = (np.genfromtxt(filename9, usecols=i, autostrip=True) for i in range(2))

fig = plt.figure(figsize=(11, 7) , dpi=130)
plt.xlabel('Time step')
plt.errorbar(x, etot, yerr=erretot, color='k', label='Energia totale')
plt.errorbar(x, epot, yerr=errepot, color='r', label='Energia potenziale')
plt.errorbar(x, ekin, yerr=errekin, color='c', label='Energia cinetica')
plt.errorbar(x, temp, yerr=errtemp, color='y', label='Temperatura')
plt.grid(True)
plt.legend(loc=0)
fig.tight_layout()
plt.draw()
plt.show()
```
