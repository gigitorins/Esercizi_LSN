```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import math
import scipy.stats
from itertools import cycle
from mpl_toolkits.mplot3d import Axes3D
import scipy
import os

currentDirectory = os.getcwd()
filename1 = "%s/nsim.dat" % currentDirectory
nsim = np.genfromtxt(filename1, usecols=0) - 1
M = 10**4                # Numero di time-steps
N = 200                  # Numero di blocchi
x = np.arange(0,nsim*M,step=10)

# Caricamento dati da file e grafici
clrlist = ['r','c','k','y']
fig = plt.figure(figsize=(11, 7) , dpi=120)
filename2 = '%s/%s' % (currentDirectory,'output_temp.dat')
y = np.genfromtxt(filename2, usecols=0, autostrip=True)
plt.plot(x, y, color=clrlist[3], label='Temperatura')
plt.xlabel('Time-steps')
plt.grid(True)
plt.legend(loc=0)
plt.title('Temperatura in un gas di LJ')
fig.tight_layout()
plt.draw()
plt.show()
```

```python
currentDirectory = os.getcwd()
#filename1 = "%s/nsim.dat" % currentDirectory
#nsim = np.genfromtxt(filename1, usecols=0) - 1
nsim=1

M = 10**4                # Numero di time-steps
N = 200                  # Numero di blocchi
x = np.arange(nsim*N)
kb = scipy.constants.Boltzmann

# Caricamento dati da file e grafici
gnames = ['Energia potenziale','Energia cinetica','Energia totale','Temperatura']
fnames = ['ave_epot.', 'ave_ekin.', 'ave_etot.', 'ave_temp.']
tnames, typenames = ['gsout','liqout','sldout'], ['Gas', 'Liquid', 'Solid']
clrlist = ['r','c','k','y']
siunits = [120.*kb,120.*kb,120.*kb,1./120.]
fig = plt.figure(figsize=(16, 10) , dpi=150)

for i in range(3):
    ax = fig.add_subplot(2, 2, i+1)
    ax.set_title('%s' % typenames[i], fontweight='bold', fontsize=12)
    for j in range(3):
        filename2 = '%s/%s%s' % (currentDirectory,fnames[j],tnames[i])
        y, yerror = (np.genfromtxt(filename2, usecols=i, autostrip=True) for i in range(2))
        y *= siunits[j]
        yerror *= siunits[j]
        if j==2:
            ax.errorbar(x, y, yerr=yerror, color=clrlist[j] , label='%s' % gnames[i])
        else:
            ax.errorbar(x, y, yerr=yerror, color=clrlist[j])
            
    ax.grid(True)
   

ax = fig.add_subplot(2, 2, 4)
ax.set_xlabel('Time-steps')

fig.legend(bbox_to_anchor=(0.25, 0.53, 0.5, 0.5), fontsize='15', loc='upper center',
           prop=dict(weight='bold'), ncol=3, mode="expand", borderaxespad=0.)
fig.suptitle('Argon MD, simulation n. 5', y=1.09, fontweight='bold', fontsize=18)
fig.tight_layout()
plt.draw()
plt.show()
#fig.savefig('argon.png', bbox_inches='tight')
```
