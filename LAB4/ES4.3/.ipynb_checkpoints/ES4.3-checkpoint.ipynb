# Esercizio 4.3 - Molecular Dynamics of Argon

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import math
import scipy.stats
from itertools import cycle
from IPython.display import clear_output
import time
import scipy
from scipy.optimize import curve_fit
import subprocess
import os

N = 100   # Numero di blocchi
gnames = ['Energia potenziale','Energia cinetica','Energia totale','Temperatura'] # Nomi dei grafici
fnames = ['ave_epot.out', 'ave_ekin.out', 'ave_etot.out', 'ave_temp.out'] # Nomi dei file di output
ylabels = [r'$E_{pot}$ [ J ]', r'$E_{kin}$ [ J ]', r'$E_{tot}$ [ J ]', r'$T$ [ K ]'] # Nomi dei file di output
kb = scipy.constants.Boltzmann
siunits = [120.*kb,120.*kb,120.*kb,120.]

def replace_line(file_name, line_num, text):
    with open(file_name, 'r') as file:
        data = file.readlines()

    data[line_num] = text

    with open(file_name, 'w') as file:
        file.writelines(data)
```

Simulazione delle tre fasi dell'argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) utilizzando un potenziale di Lennard-Jones.

Le condizioni utilizzate, presenti rispettivamente nei file <font face="Courier">argon.solid</font>, <font face="Courier">argon.liquid</font> e <font face="Courier">argon.gas</font> sono:

1) fase solida: &nbsp;&nbsp;&nbsp;&nbsp; $\rho^* = 1.1$,&nbsp;&nbsp;&nbsp;&nbsp; $T^* = 0.8$, &nbsp;&nbsp;$r_c = 2.2$

2) fase liquida: &nbsp;&nbsp;&nbsp; $\rho^* = 0.8$,&nbsp;&nbsp;&nbsp;&nbsp; $T^* = 1.1$, &nbsp;&nbsp;$r_c = 2.5$

3) fase gassosa: &nbsp;$\rho^* = 0.05$, &nbsp;&nbsp;$T^* = 1.2$,&nbsp;&nbsp; $r_c = 5.0$

Il numero di particelle è fissato a $n_\text{part}=108$, l'intervallo temporale utilizzato è $dt=5*10^{-4}$ (circa un decimo di femtometro in unità SI) con $10^4$ steps temporali.

I grafici mostrano $E/N$, $U/N$, $K/N$ e $T$ con le rispettive incertezze in unità SI. Per passare da unità di LJ a unita SI si applica:

- $E = \epsilon E^*,\sigma_E = \frac{\partial E}{\partial E^*}\sigma_E^*=\epsilon\sigma_E^*$ (per energia totale, cinetica e potenziale);

- $T=\frac{\epsilon}{k_b}T^*,\sigma_T=\frac{\epsilon}{k_b}\sigma_T^*$

Per ogni fase ho svolto aclune simulazioni iniziali di equilibrazione; ho usato 1000 steps rispetto ai 10000 steps della simulazione vera e propria per il calcolo del valore medio e dell'incertezza delle grandezze. Le simulazioni iniziali servono infatti solo a equilibrare il sistema riscalando le velocità, per ottenere una temperatura vicina a quella target.

## --------------------------------------------- SOLIDO ---------------------------------------------

### Equilibrazione e simulazione

```python
state = 'solid'
nsim = 6  # Numero di simulazioni (5 per equilibrazione + 1 per simulazione)
# Elimino files precedenti, imposto config.0 e input.dat
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
subprocess.run("cp config.fcc config.0", shell=True, stdout=subprocess.PIPE)
subprocess.run("make cp"+state, shell=True, stdout=subprocess.PIPE)

# Eseguo simulazioni
for i in range(nsim-1):
    arguments = str(i)+" "+str(100)+" "+str(1000)
    subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
```

Il primo plot mostra che la temperatura si stabilizza attorno alla temperatura desiderata $T^*=0.8$ dopo aver fatto ripartire la simulazione 5 volte.

Si può osservare come l'energia totale mostri una buona convergenza e venga conservata durante la simulazione. La temperatura si assesta attorno al valore di $97 K$, vicina al valore atteso $T=\epsilon/k_bT^*=120*0.8 = 96 K$.

```python
fig = plt.figure(figsize=(16, 10))
y = np.genfromtxt('output_temp.dat', usecols=0, autostrip=True)
x = np.arange(0,len(y))*10
plt.plot(x, y)
plt.hlines(0.8, 0, len(y)*10, color='r', linestyles='solid', label='T*')
plt.xlabel('Time steps', fontsize=14)
plt.ylabel('Temperatura [L-J units]', fontsize=14)
plt.grid(True)
plt.show()
```

### Simulazione

```python
subprocess.run("make esegui ARGS="+str(nsim), shell=True)
```

```python
# Grafici per l'ultima run (post-equilibrazione)
fig = plt.figure(figsize=(16, 12))
xs = np.arange(0,N)

for i in range(4):
    ax = fig.add_subplot(2, 2, i+1)
    ys, yerrors = (np.genfromtxt(fnames[i], usecols=j, autostrip=True) for j in range(2))
    ys *= siunits[i]
    yerrors *= siunits[i]
    ax.errorbar(xs, ys[-N:], yerr=yerrors[-N:])
    ax.set_xlabel('Numero di blocchi',fontsize=14)
    ax.set_ylabel(ylabels[i],fontsize=14)
    ax.set_title(gnames[i], fontweight='bold', fontsize=16)
    ax.grid(True)

fig.tight_layout()
plt.show()
```

## --------------------------------------------- LIQUIDO ---------------------------------------------

### Equilibrazione e simulazione

```python
state = 'liquid'
nsim = 10  # Numero di simulazioni (7 per equilibrazione + 1 per simulazione)
# Elimino files precedenti, imposto config.0 e input.dat
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
subprocess.run("cp config.fcc config.0", shell=True, stdout=subprocess.PIPE)
subprocess.run("make cp"+state, shell=True, stdout=subprocess.PIPE)

# Eseguo simulazioni
for i in range(nsim-1):
    arguments = str(i)+" "+str(100)+" "+str(1000)
    subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
```

Il primo plot mostra che la temperatura si stabilizza attorno alla temperatura desiderata $T^*=1.1$ dopo 8000 steps. Si può osservare come le energie oscillino di più rispetto alla fase solida, ma l'$E_{tot}$ è comunque conservata durante la simulazione. La temperatura si assesta attorno al valore di $129 K$, vicina al valore atteso $T=\epsilon/k_bT^*=120*1.1 = 132 K$.

```python
fig = plt.figure(figsize=(16, 10))
y = np.genfromtxt('output_temp.dat', usecols=0, autostrip=True)
x = np.arange(0,len(y))*10
plt.plot(x, y)
plt.hlines(1.1, 0, len(y)*10, color='r', linestyles='solid', label='T*')
plt.xlabel('Time steps', fontsize=14)
plt.ylabel('Temperatura [L-J units]', fontsize=14)
plt.grid(True)
plt.show()
```

### Simulazione

```python
subprocess.run("make esegui ARGS="+str(nsim), shell=True)
```

```python
# Grafici per l'ultima run (post-equilibrazione)
fig = plt.figure(figsize=(16, 12))
xl = np.arange(0,N)

for i in range(4):
    ax = fig.add_subplot(2, 2, i+1)
    yl, yerrorl = (np.genfromtxt(fnames[i], usecols=j, autostrip=True) for j in range(2))
    yl *= siunits[i]
    yerrorl *= siunits[i]
    ax.errorbar(xl, yl[-N:], yerr=yerrorl[-N:])
    ax.set_xlabel('Numero di blocchi',fontsize=14)
    ax.set_ylabel(ylabels[i],fontsize=14)
    ax.set_title(gnames[i], fontweight='bold', fontsize=16)
    ax.grid(True)

fig.tight_layout()
plt.show()
```

## ----------------------------------------------- GAS -----------------------------------------------

### Equilibrazione e simulazione

```python
state = 'gas'
nsim = 7  # Numero di simulazioni (7 per equilibrazione + 1 per simulazione)
# Elimino files precedenti, imposto config.0 e input.dat
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
subprocess.run("cp config.fcc config.0", shell=True, stdout=subprocess.PIPE)
subprocess.run("make cp"+state, shell=True, stdout=subprocess.PIPE)

# Eseguo simulazioni
for i in range(nsim):
    arguments = str(i)+" "+str(100)+" "+str(5000)
    subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
```

La fase gassosa è la più difficile da equilibrare; ho utilizzato 6 simulazioni di equilibrazione da 5000 steps. La temperatura varia fortemente nel corso delle simulazioni ma si stabilizza infine attorno alla temperatura desiderata $T^*=1.2$, pur discostandosi maggiormente rispetto all'equilibrazione delle fasi solida e liquida. Le oscilazioni sono molto più marcate rispetto alle altre due fasi; la temperatura misurata si assesta attorno al valore di $145 K$, vicinissima al valore atteso $T=\epsilon/k_bT^*=120*1.2 = 144 K$.

```python
fig = plt.figure(figsize=(16, 10))
y = np.genfromtxt('output_temp.dat', usecols=0, autostrip=True)
x = np.arange(0,len(y))*10
plt.plot(x, y)
plt.hlines(1.2, 0, len(y)*10, color='r', linestyles='solid', label='T*')
plt.xlabel('Time steps', fontsize=14)
plt.ylabel('Temperatura [L-J units]', fontsize=14)
plt.grid(True)
plt.show()
```
