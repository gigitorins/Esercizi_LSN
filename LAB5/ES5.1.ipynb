# Esercizio 5.1 - M(RT)$^2$

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import math
import scipy.stats
from itertools import cycle
from mpl_toolkits.mplot3d import Axes3D
import scipy
import os
import subprocess

M = 10**6              # Numero di lanci
N = 100                # Numero di blocchi
L = int(M/N)           # Numero di lanci in ogni blocco (M deve essere multiplo di N)
bohr0 = 0.0529 * 10**(-9)

fnames = ['file1.txt', 'file2.txt']
rvalues = [3./2, 5.]
gnames = [r'$\psi_{1,1,0}$', r'$\psi_{2,1,0}$']
names = [r'$\psi_{1,1,0}$', r'$\psi_{2,1,0}$']
```

Utilizziamo l'algoritmo di Metropolis per campionare la distribuzione di probabilità della posizione dell'elettrone di un atomo di idrogeno. La PDF è data dal modulo quadro della funzione d'onda; in coordinate cartesiane per l'**orbitale 1s** si ha:

$$p_{1,0,0}\left(x,y,z\right)=\frac{a_0^{-3}}{\pi} \exp\left({-\frac{2}{a_0}\sqrt{x^2+y^2+z^2}}\right)$$

Per l'**orbitale 2pz** invece:

$$p_{2,1,0}\left(x,y,z\right)=\frac{a_0^{-5}}{32\pi}z^2\exp\left({-\frac{1}{a_0}\sqrt{x^2+y^2+z^2}}\right)$$


Ho utilizzato inizialmente una probabilità di transizione uniforme $T\left(\vec x^*| \vec y\right)$; ogni componente di $\vec x^*$ viene estratta uniformemente tra $y_i - \Delta$ e $y_i + \Delta$ dove $y_i$ indica una componente del vettore $\vec y$ e $\Delta$ è la larghezza dell'intervallo di estrazione.

```python
x1, y1, z1 = (np.genfromtxt("fileR1.txt", usecols=i, autostrip=True) for i in range(3))
coord = [x1/bohr0,y1/bohr0,z1/bohr0]
fig = plt.figure(figsize=(9,9))
ax = fig.add_subplot(projection='3d')
ax.view_init(5, 30)
ax.grid(True)
ax.xaxis.pane.fill = ax.yaxis.pane.fill = ax.zaxis.pane.fill = False
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.scatter(coord[0], coord[1], coord[2], c=coord[2], marker='.')
ax.set_title(r'$\psi_{1,1,0}$', fontsize=15)
plt.show()

fig = plt.figure(figsize=(14, 7) )
ax = fig.add_subplot()
x = np.arange(N)
r, err = (np.genfromtxt('file1.txt', usecols=i, autostrip=True) for i in range(2))
ax.errorbar(x, r/bohr0, yerr = err/bohr0)
ax.hlines(rvalues[0], 0, (N-1), colors='r', linestyles='solid', label='Expectation values for the radius')
ax.set_xlabel('N.° blocchi',  fontsize=14)
ax.set_ylabel('<r>',  fontsize=14)
ax.grid(True)
plt.show()
```

Uno degli accorgimenti iniziali è fissare il punto di partenza dell'algoritmo. Facendo partire la simulazione da un punto lontano dall'origine si nota che la PDF campionata dall'algoritmo per l'orbitale $1s$ non assomiglia a quella tipica, e che il valor medio del raggio impiega circa 50 blocchi a stabilizzarsi attorno al valore atteso.
Scegliendo il punto di partenza vicino all'origine la PDF ottenuta sarà simile a un orbitale $1s$ senza dover escludere i punti campionati all'inizio dall'algoritmo di Metropolis. Lo stesso ragionamento è stato fatto per l'orbitale $2pz$.

I successivi punti dell'esperimento verranno generati utilizzando $T\left(\vec x^*| \vec x_n\right)$; la probabilità di accettare la mossa $A\left(\vec x^*| \vec x_n\right) = \min\left(1,p\left(\vec x^*\right)/p\left(\vec x_n\right)\right)$ ha questa forma perchè $T$ è simmetrica (ovvero $T\left(\vec x^*| \vec x_n\right)=T\left(\vec x_n| \vec x^*\right)$).

L'algoritmo decide se accettare una mossa estraendo casualmente un numero r in $\left[0,1\right]$:
- se $r \leq A\left(\vec x^*| \vec x_n\right)$ allora accetta la mossa e $x_{n+1}=x^*$;
- se $r \geq A\left(\vec x^*| \vec y\right)$ la rifiuta e $x_{n+1} = x_n$.

La scelta del passo $\Delta$ dovrebbe rispettare la regola empirica che fissa a 50% il rate di accettazione delle mosse proposte. In questo caso il valore ideale è $\Delta=1.22$, per il quale $\langle A\rangle\simeq50.25\%$

Utilizzando il data blocking si può lanciare l'algoritmo $M$ volte, dividendo gli $M$ lanci in $N$ blocchi per ottenere il valore medio del raggio $\langle r\rangle$ e la sua incertezza.
Il numero di blocchi dovrebbe essere scelto in modo che il rate di accettazione $A$ per ogni blocco non sia eccessivamente variabile nel corso della simulazione e che i punti non siano troppo correlati tra loro (si dovrebbe quindi avere $A \simeq50\%$ in tutti i blocchi). La scelta migliore è $N=100$.

Per l'orbitale $2pz$ la simulazione converge verso il valore teorico $\langle r\rangle = 5 a_0$ con una scelta ottimale del passo pari a $\Delta=2.98$ e numero di blocchi $N=100$.

**I valori medi del raggio $\langle r \rangle$ riportati nei grafici sono in unità di raggi di Bohr $a_0$.**

## Grafici per $T(\vec{x}|\vec{y})$ gaussiana

Se si utilizza come probabilità di transizione $T$ una PDF Gaussiana per ogni coordinata, il risultato per l'orbitale $1s$ è praticamente identico al caso precedente, mentre la stima di $\langle r \rangle$ per l'orbitale $2pz$ si avvicina di più al valore atteso. Anche con $T$ gaussiana la scelta ottimale per il numero di blocchi è $N=100$. La PDF utilizzata è del tipo:

$$T\left((x^*)_i|(x_n)_i\right)=\mathcal{N}\left( (x_n)_i,\delta\right)$$

Le scelte per $\Delta$ in questo caso sono:
- Orbitale $1s$: &nbsp;&nbsp;&nbsp;&nbsp;$\Delta = 0.76$ &nbsp; $\rightarrow \langle A\rangle \simeq50.02\%$;
- Orbitale $2pz$: &nbsp;&nbsp;$\Delta = 1.875 \rightarrow \langle A\rangle \simeq50.08\%$

```python
fig = plt.figure(figsize=(12, 12) )
fnames = ['file3.txt', 'file4.txt']
x = np.arange(N)

for i in range(2):
    ax = fig.add_subplot(2, 1, i+1)
    r, err = (np.genfromtxt(fnames[i], usecols=j, autostrip=True) for j in range(2))
    ax.errorbar(x, r/bohr0, yerr = err/bohr0)
    ax.hlines(rvalues[i], 0, (N-1), colors='r', linestyles='solid', label='Expectation values for the radius')
    ax.set_xlabel('N.° blocchi',  fontsize=14)
    ax.set_ylabel('<r>',  fontsize=14)
    ax.set_title(gnames[i], fontsize=16)
    ax.grid(True)
    
fig.tight_layout()
plt.show()
```
