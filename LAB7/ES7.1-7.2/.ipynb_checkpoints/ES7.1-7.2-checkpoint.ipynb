# -------------------------------------------- Esercizio 7.1 --------------------------------------------

Equilibro ogni simulazione eseguendo 10000 steps, successivamente genero i valori istantanei di $U/N$ e $P$
per $M=5\times 10^5$ MC steps (con una accettazione fissata al 50%). Le condizioni termodinamiche considerate sono:
- liquido: &nbsp;$\rho^\star = 0.8$, &nbsp;&nbsp;&nbsp;&nbsp;$T^\star = 1.1$&nbsp; (raggio di cut-off: $r_c = 2.5$)
- gas: &nbsp;&nbsp; &nbsp; $\rho^\star = 0.05$, &nbsp;&nbsp;$T^\star = 1.2$&nbsp; (raggio di cut-off: $r_c = 5.0$)
- solido:&nbsp; $\rho^\star = 1.1$, &nbsp;&nbsp;&nbsp;&nbsp;$T^\star = 0.8$&nbsp; (raggio di cut-off: $r_c = 2.2$)

Dopo le run vengono calcolate la funzione di autocorrelazione per $U/N$ e $P$ in modo da stabilire la grandezza minima dei blocchi tale da evitare punti fortemente correlati tra loro. Gli stessi dati sono utilizzati anche per calcolare l'incertezza statistica dei valori di aspettazione di $U/N$ e $P$ in funzione della dimensione dei blocchi nel data blocking.

**I valori di energia interna e pressione mostrati nei grafici sono in unit√† di Lennard-Jones.**

## Funzione di autocorrelazione

```python
start = time.time()
autocorrE, autocorrP = [], []
npoints = 300

for k in range(npoints):
    clear_output(wait=True)
    print('-- Progression: ', int((k+1)/npoints*100), "% \r")
    epot1 = yvect[0][:len(yvect[0]) - k]
    epotk = yvect[0][k:]
    pres1 = yvect[1][:len(yvect[1]) - k]
    presk = yvect[1][k:]
    n = len(epot1)
    autocorrE.append((sum(epot1 * epotk)/n - sum(epot1) / n * sum(epotk) / n)/np.std(yvect[0])**2)
    autocorrP.append((sum(pres1 * presk)/n - sum(pres1) / n * sum(presk) / n)/np.std(yvect[1])**2)
    
end = time.time()
autocorr = [autocorrE, autocorrP]
print('-- Tempo:', int(end-start),'sec')
```

```python
fig = plt.figure(figsize=(14, 7))
gnames = ['Autocorrelazione per U/N','Autocorrelazione per P'] # Nomi grafici
x = np.arange(npoints)

def test_func(x, a):
    return np.exp(-a*x)

params_epot, params_cov_epot = curve_fit(test_func, x, autocorrE)
params_pres, params_cov_pres = curve_fit(test_func, x, autocorrP)
paramstot = [params_epot, params_pres]

for i in range(2):
    ax = fig.add_subplot(1, 2, i+1)
    ax.plot(autocorr[i], color='blue', label = 'Autocorrelazione')
    ax.plot(x, test_func(x,paramstot[i]), color='red', label='Fit')
    ax.grid(True)    
    ax.set_xlabel('Steps', fontsize=13)
    ax.set_title(gnames[i], fontsize=15)
    ax.legend(loc=0)
    
fig.tight_layout()
plt.show()
```

Dal grafico si pu√≤ stabilire che i blocchi devono essere larghi almeno 200 steps.


## Incertezza statistica nel data blocking

Utilizzo i valori istantanei dell'energia come caso di studio. Per la stima dell'incertezza statistica di $\langle U/N \rangle$ e $\langle P \rangle$ sono stati considerati blocchi di dimensioni variabili tra $L=M/N=10$ a $L=5\times 10^3$, con $N$ numero di blocchi. Il grafico dell'autocorrelazione mostra che per valori $ > 200$ steps si perde completamente la correlazione tra i dati. La figura seguente ci permette di osservare che l'errore satura dopo i $500$ dati per blocco: questo sar√† quindi il valore da utilizzare.

```python
L = np.arange(100)*50 + 10
M = len(yvect[0])
Nnew = M/L
start = time.time()
final_errors = np.zeros(len(L))
nn = 0

for l in L:
    clear_output(wait=True)
    print('-- Progression: ', int((l+1)/max(L)*100), "% \r")
    Nnew = int(M/l)
    avg, err = (np.zeros(Nnew) for i in range(2))
    for i in range(Nnew):
        for j in range(l):
            avg[i] += (yvect[0][j+l*i])/l
        accumulate, accumulate_2 = (0 for i in range(2))
        for j in range(i+1):
            accumulate += avg[j]
            accumulate_2 += avg[j]**2
        glob_avg = accumulate/(i+1)
        glob_avg_2 = accumulate_2/(i+1)
        err = np.sqrt((glob_avg_2 - glob_avg**2)/(i+1))
    final_errors[nn] = err
    nn += 1
    
end = time.time()
print('-- Tempo: ', (end-start), 'sec')
```

```python
fig = plt.figure(figsize=(14, 7) , dpi=130)
plt.plot(L, final_errors)
plt.axvline(500, color='red')
plt.grid(True)  
plt.xlabel("Number of steps per block", fontsize=13)
plt.ylabel("Error", fontsize=13)
plt.show()
```

# ---------------------------------------- Esercizio 7.2 ----------------------------------------

Sono riportati i grafici della funzione di distribuzione radiale $g(r)$ in funzione della distanza radiale tra particelle con $r \in \left[ 0; L/2\right]$, con relativa incertezza statistica.

$g\left(\vec r_1, \vec r_2\right)$ √® definita come il rapporto tra la densit√† a due particelle del sistema studiato e la densit√† a due particelle del gas ideale, ovvero
$$g\left(\vec r_1, \vec r_2\right)=\frac{\rho^{(2)}\left(\vec r_1,\vec r_2\right)}{\rho^{(1)}\left(\vec r_1\right)\rho^{(1)}\left(\vec r_2\right)}$$

Il calcolo corrisponde a creare un istogramma e ad aggiungere due unit√† nel bin corrispondente alla distanza $r$ ogni volta che due particelle sono a una distanza tra $r$ e $r+dr$. Dopo aver riempito l'istogramma, si normalizza dividendo ogni bin per $\rho n_\text{part}\Delta V(r)$. Il numero di bin scelto √® 100.

```python
n_blocks = 10
n_steps = 1000
start = time.time()
```
