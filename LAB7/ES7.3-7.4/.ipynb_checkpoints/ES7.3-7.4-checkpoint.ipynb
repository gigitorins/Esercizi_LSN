# Esercizio 7.3 / 7.4 - Monte Carlo NVT simulation of Argon

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import math
import scipy.stats
from itertools import cycle
from IPython.display import clear_output
import time
import scipy
from scipy.optimize import curve_fit
import subprocess
import os
```

Equilibro e svolgo delle simulazioni MC NVT di un modello di Lennard-Jones dell'Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu)  
nelle seguenti condizioni termodinamiche:
- liquido: &nbsp;$\rho^\star = 0.8$, &nbsp;&nbsp;&nbsp;&nbsp;$T^\star = 1.1$&nbsp; (raggio di cut-off: $r_c = 2.5$)
- gas: &nbsp;&nbsp; &nbsp; $\rho^\star = 0.05$, &nbsp;&nbsp;$T^\star = 1.2$&nbsp; (raggio di cut-off: $r_c = 5.0$)
- solido:&nbsp; $\rho^\star = 1.1$, &nbsp;&nbsp;&nbsp;&nbsp;$T^\star = 0.8$&nbsp; (raggio di cut-off: $r_c = 2.2$)

Nei grafici sono riportati i valori medi di $U/N$ e $P$ con le loro incertezze in funzione del numero di blocchi. Il numero di blocchi e il numero di steps per blocco utilizzati nella simulazione di ogni fase sono quelli ricavati negli esercizi <a href="../ES7.1-7.2/ES7.1-7.2.ipynb">7.1-7.2</a>, che minimizzano la correlazione tra i punti e l'incertezza dei valori medi.

Viene infine confrontata la funzione di distribuzione radiale $g(r)$ generata con le simulazioni MC NVT con quella ottenuta tramite il codice Molecular Dynamics NVE in condizioni termodinamiche simili.

## Variabili e parametri 

```python
CWD = os.getcwd() # Current work directory
begin = time.time()
fnames = ['output.epot.0', 'output.pres.0'] # Nomi dei file di output
gnames = ['Energia potenziale','Pressione'] # Nomi grafici
labnames = ['U/N [J]', 'P [atm]'] # Labels asse y
colorlist = ['blue', 'red']
kb = scipy.constants.Boltzmann
epsi = 120.*kb
sigma3 = (0.34*(10**(-9)))**(3)
siunits = [epsi, epsi/sigma3/(101325.)]
os.chdir("%s" % CWD)
start = time.time()
```

## -------------------------------------------------------- SOLIDO --------------------------------------------------------

```python
state = 'solid'
n_blocks = 100
n_steps = 300
subprocess.run("cp config.fcc config.0", shell=True, stdout=subprocess.PIPE)
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE) # Pulisco i file main.exe, *.o
print('Inizio equilibrazione')
arguments = "input."+str(state)+" "+str(10)+" "+str(300)
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
print('Fine equilibrazione')
subprocess.run("cp config.final config.0", shell=True, stdout=subprocess.PIPE)
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE) # Pulisco i file main.exe, *.o
arguments = "input."+str(state)+" "+str(n_blocks)+" "+str(n_steps)
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
print('Fine simulazione')

# Dati g(r)
rvecs, ave_solid, err_solid = (np.genfromtxt('output.gave.0', usecols=kk, autostrip=True) for kk in range(3))
```

```python
fig = plt.figure(figsize=(14, 6))

for i in range(2):
    ax = fig.add_subplot(1, 2, i+1)
    x, appo, y, yerror = (np.genfromtxt(fnames[i], usecols=kk, autostrip=True) for kk in range(4))
    y *= siunits[i]
    yerror *= siunits[i]
    ax.errorbar(x, y, yerr=yerror, color=colorlist[i])
    ax.set_xlabel('Numero blocchi', fontsize=13)
    ax.set_ylabel(labnames[i], fontsize=13)
    ax.set_title(gnames[i], fontsize=15)
    ax.grid(True)
    
fig.tight_layout()
plt.show()
```

```python
state = 'liquid'
n_blocks = 100
n_steps = 500
subprocess.run("cp config.fcc config.0", shell=True, stdout=subprocess.PIPE)
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE) # Pulisco i file main.exe, *.o
print('Inizio equilibrazione')
arguments = "input."+str(state)+" "+str(10)+" "+str(500)
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
print('Fine equilibrazione')
subprocess.run("cp config.final config.0", shell=True, stdout=subprocess.PIPE)
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE) # Pulisco i file main.exe, *.o
arguments = "input."+str(state)+" "+str(n_blocks)+" "+str(n_steps)
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
print('Fine simulazione')

# Dati g(r)
rvecl, ave_liquid, err_liquid = (np.genfromtxt('output.gave.0', usecols=kk, autostrip=True) for kk in range(3))
```

```python
fig = plt.figure(figsize=(14, 6))

for i in range(2):
    ax = fig.add_subplot(1, 2, i+1)
    x, appo, y, yerror = (np.genfromtxt(fnames[i], usecols=kk, autostrip=True) for kk in range(4))
    y *= siunits[i]
    yerror *= siunits[i]
    ax.errorbar(x, y, yerr=yerror, color=colorlist[i])
    ax.set_xlabel('Numero blocchi', fontsize=13)
    ax.set_ylabel(labnames[i], fontsize=13)
    ax.set_title(gnames[i], fontsize=15)
    ax.grid(True)
    
fig.tight_layout()
plt.show()
```

# Calcolo della funzione di distribuzione radiale in MD

Sono mostrate le funzioni di distribuzione radiale $g(r)$ per gli ensemble canonico e microcanonico. Per la simulazione della fase solida le due curve sono praticamente coincidenti; l'accordo tra le due funzioni è prevedibile, dato stiamo simulando lo stesso sistema in un caso a temperatura fissata (NVT) e nell'altro caso a energia fissata (NVE). Nel limite termodinamico, i grafici di $g(r)$ dei due casi siano equivalenti.

Per ogni fase simulata con il codice MD NVE, vengono utilizzati lo stesso numero di blocchi e di steps per blocco delle relative simulazioni MC NVT.

I grafici di $g(r)$ generati in MD per le fasi liquida e gassosa si discostano leggermente da quelle NVT in corrispondenza del picco principale (effetto più marcato per il gas), ma l'andamento qualitativo e l'accordo per il resto del range di valori sono ottimi. Mentre per la fase solida la simulazione è stata fatta ripartire 10 volte, per le altre fasi sono serviti rispettivamente 15 e 30 restart per ottenere grafici di $g(r)$ accurati; ciò è in accordo con la necessità di far ripartire più volte la simulazione per queste due fasi rispetto alla fase solida nell'<a href="../../LAB4/ES4.3/ES4.3.ipynb">Esercitazione 4</a>.

```python
os.chdir("%s/MD" % CWD) # Subdirectory MD
start = time.time()
print("%s/MD" % CWD)
```

```python
# Solid
state = 'solid'
nsim = 10    # Numero di simulazioni
N = 100      # Numero di blocchi
nsteps = 300 # Steps per blocco
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
subprocess.run("cp config.fcc config.0", shell=True, stdout=subprocess.PIPE)
subprocess.run("make cp"+state, shell=True, stdout=subprocess.PIPE)

for i in range(nsim):
    arguments = str(i)+" "+str(10)+" "+str(100)
    subprocess.run("make esegui ARGS='%s'" % arguments, shell=True)

arguments = str(nsim)+" "+str(N)+" "+str(nsteps)
subprocess.run("make esegui ARGS='%s'" % arguments, shell=True)
rvecs1, ave_solid1, err_solid1 = (np.genfromtxt('output.gave.0', usecols=kk, autostrip=True) for kk in range(3))
```

```python
# Liquid
state = 'liquid'
nsim = 15    # Numero di simulazioni
N = 100      # Numero di blocchi
nsteps = 500 # Steps per blocco
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
subprocess.run("cp config.fcc config.0", shell=True, stdout=subprocess.PIPE)
subprocess.run("make cp"+state, shell=True, stdout=subprocess.PIPE)

for i in range(nsim):
    arguments = str(i)+" "+str(100)+" "+str(100)
    subprocess.run("make esegui ARGS='%s'" % arguments, shell=True)

arguments = str(nsim)+" "+str(N)+" "+str(nsteps)
subprocess.run("make esegui ARGS='%s'" % arguments, shell=True)
rvecl1, ave_liquid1, err_liquid1 = (np.genfromtxt('output.gave.0', usecols=kk, autostrip=True) for kk in range(3))
```

```python
# Gas
state = 'gas'
nsim = 20    # Numero di simulazioni
N = 100      # Numero di blocchi
nsteps = 300 # Steps per blocco
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
subprocess.run("cp config.fcc config.0", shell=True, stdout=subprocess.PIPE)
subprocess.run("make cp"+state, shell=True, stdout=subprocess.PIPE)

for i in range(nsim):
    arguments = str(i)+" "+str(10)+" "+str(100)
    subprocess.run("make esegui ARGS='%s'" % arguments, shell=True)

arguments = str(nsim)+" "+str(N)+" "+str(nsteps)
subprocess.run("make esegui ARGS='%s'" % arguments, shell=True)
rvecg1, ave_gas1, err_gas1 = (np.genfromtxt('output.gave.0', usecols=kk, autostrip=True) for kk in range(3))

for i in range(10):
    arguments = str(nsim+1+i)+" "+str(10)+" "+str(100)
    subprocess.run("make esegui ARGS='%s'" % arguments, shell=True)

arguments = str(nsim+11)+" "+str(N)+" "+str(nsteps)
subprocess.run("make esegui ARGS='%s'" % arguments, shell=True)
rvecg2, ave_gas2, err_gas2 = (np.genfromtxt('output.gave.0', usecols=kk, autostrip=True) for kk in range(3))
end = time.time()
print('Computational time: ', int(end-start), 'sec')
```

## Funzione di distribuzione radiale g(r)

```python
os.chdir("%s" % CWD)
rtotal = [rvecs, rvecl, rvecg]
ave_total = [ave_solid, ave_liquid, ave_gas]
err_total = [err_solid, err_liquid, err_gas]
rtotal1 = [rvecs1, rvecl1, rvecg1]
ave_total1 = [ave_solid1, ave_liquid1, ave_gas1]
err_total1 = [err_solid1, err_liquid1, err_gas1]
graphnames = ['Solido', 'Liquido', 'Gas']
fig = plt.figure(figsize=(14, 15) , dpi=130)

for i in range(3):
    ax = fig.add_subplot(3, 1, i+1)
    ax.errorbar(rtotal[i], ave_total[i], yerr=err_total[i], color="b", ecolor="k", label = 'NVT')
    ax.errorbar(rtotal1[i][-100:], ave_total1[i][-100:], yerr=err_total1[i][-100:],
                color="r", ecolor="k", label = 'MD')
    if i==2:
        ax.errorbar(rvecg2[-100:], ave_gas2[-100:], yerr=err_gas2[-100:],
                    color="orange", ecolor="k", label = 'MD, 30 ripartenze')
    ax.set_xlabel('r/σ', fontsize=15)
    ax.set_ylabel('g(r)', fontsize=15)
    ax.set_title(graphnames[i], fontsize=16)
    ax.legend(loc=0)
    ax.grid(True) 
    
fig.tight_layout()
plt.show()
```
