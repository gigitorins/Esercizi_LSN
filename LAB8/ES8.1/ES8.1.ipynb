L'esercizio consiste nell'implementare un codice Monte Carlo per stimare il valore medio dell'Hamiltoniana

$$\hat H=-\frac{\hbar}{2}\frac{\partial^2}{\partial x^2}+x^4+\frac{5}{2}x^2$$

utilizzando una funzione d'onda di prova il cui modulo quadro è:

$$|\Psi_T^{\sigma,\mu}(x)|^2 =|\mathcal{N}|^2 \left(e^{-\frac{(x-\mu)^2}{2\sigma^2}}+e^{-\frac{(x+\mu)^2}{2\sigma^2}}\right)^2$$

Il codice Monte Carlo variazionale utilizza l'algoritmo di Metropolis per campionare la funzione d'onda test usando una probabilità di transizione uniforme $T(x_{new}|x_{old})$.
Data un'energia locale pari a

$$ E_\text{loc}(x) = \frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)} {\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)} $$

si può ottenere il valore di aspettazione dell'Hamiltoniana come:

$$ \langle\hat{H}\rangle_T=\frac{1}{M} \sum_{i=1}^M E_\text{loc}(x_i)$$

e la relativa incertezza statistica usando il data blocking, come mostrato nel primo grafico. Nel secondo grafico viene mostrata la funzione d'onda campionata $|\Psi_T(x)|^2$, ottenuta riempendo un istogramma con le configurazioni ottenute, e viene confrontata con la curva analitica.

## Valori di aspettazione teorici

```python
def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar, m, a = 1, 1, 10
N = 1000 # number of iterations
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0]   # Step size
V = Vpot(x)
CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V) # Hamiltonian matrix
E, psi = np.linalg.eigh(H) # Compute eigenvectors and their eigenvalues
psi = np.transpose(psi) # Take the transpose & normalize
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])
```
