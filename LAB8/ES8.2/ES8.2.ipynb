## Valori di aspettazione teorici

```python
def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar, m, a = 1, 1, 10
N = 1000 # number of iterations
x_num = np.linspace(-a/2, a/2, N)
dx = x_num[1] - x_num[0] # Step size
V = Vpot(x_num)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2
CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V) # Hamiltonian matrix
E,psi = np.linalg.eigh(H) # Compute eigenvectors and their eigenvalues
psi = np.transpose(psi) # Take the transpose & normalize
psi = psi/np.sqrt(dx)
```

## Minimizzazione di $\langle {\hat H} \rangle_T$

Ho utilizzato una sorta di algoritmo di Metropolis multivariato per trovare i parametri $\sigma$ e $\mu$
che minimizzano $\langle {\hat H} \rangle_T$:

$$
\langle {\hat H} \rangle_T = \frac{\langle \Psi_T^{\sigma,\mu}| {\hat H} | \Psi_T^{\sigma,\mu} \rangle}
{\langle \Psi_T^{\sigma,\mu} | \Psi_T^{\sigma,\mu} \rangle} \ge E_0 =
\frac{\langle \Psi_0| {\hat H} | \Psi_0 \rangle}
{\langle \Psi_0 | \Psi_0 \rangle}
$$

Ho riportato nel primo grafico la stima di $\langle {\hat H} \rangle_T$ e della sua incertezza statistica in funzione del numero di blocchi per tale set di parametri. Nel secondo grafico è invece presente la
$|\Psi_T(x)|^2$ campionata, confrontata con la curva analitica e la soluzione numerica ottenuta trasformando l'equazione di Schrodinger in un'equazione matriciale.

```python
fnames = ['output.hamilton.0', 'output.histo.0'] # Nomi dei file di output
gnames = ['Hamiltoniana','Istogramma'] # Nomi dei grafici
xlabnames = ['Numero blocchi', 'N'] # Labels asse x
ylabnames = ['H', 'P'] # Labels asse y
```

```python
# Metropolis per mu e sigma
mu = 0.78
sigma = 0.6
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
arguments = str(sigma)+" "+str(mu)
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
y = np.genfromtxt('output.hamilton.0', usecols=2, autostrip=True)
min_ene = y[-1]
mu1, sigma1 = mu, sigma
f_ene = 0
start = time.time()
accepted, attempted = 0,0

for ii in range(100):
    mu1 = random.gauss(mu,0.03)
    subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
    arguments = str(sigma1)+" "+str(mu1)
    subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
    y = np.genfromtxt('output.hamilton.0', usecols=2, autostrip=True)
    f_ene = y[-1]
    attempted+=1
    if(f_ene < min_ene):
        mu = mu1
        min_ene = f_ene
        accepted+=1
    sigma1 = random.gauss(sigma,0.03)
    subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
    arguments = str(sigma1)+" "+str(mu1)
    subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
    y = np.genfromtxt('output.hamilton.0', usecols=2, autostrip=True)
    f_ene = y[-1]
    attempted+=1
    if(f_ene < min_ene):
        sigma = sigma1
        min_ene = f_ene
        accepted+=1
    sys.stdout.write("\r-- %s || μ: %f   σ: %f   E_min: %f" % (ii+1,mu,sigma,min_ene))
    sys.stdout.flush()

end = time.time()
print('\n-- Tempo computazionale: ',end-start,'sec')
print('\n-- Rate di accettazione: ',accepted*100./attempted)
print('\nμ = %f     σ = %f ' % (mu,sigma))
```

## $|\Psi_T(x)|^2$

Nel grafico sotto si osserva che l'istogramma prodotto dal codice segue la curva analitica di $|\Psi_T^{\sigma,\mu}(x)|^2$; tra l'istogramma e la curva calcolata numericamente si può osservare che c'è accordo ma la corrispondenza non è buona come la curva analitica.

```python
def wave_function_sq(x, mu, sigma):
    a = np.exp(-(x-mu)**2./(2.*sigma**2.)) + np.exp(-(x+mu)**2./(2.*sigma**2.)) 
    b = np.sqrt(2.*np.sqrt(np.pi)*sigma*(1. + np.exp(-(mu/sigma)**2.)))
    return  a/b  
```

```python
x, appo, y, yerror = (np.genfromtxt(fnames[1], usecols=kk, autostrip=True) for kk in range(4))
y_th = (wave_function_sq(x, mu, sigma))**2

fig = plt.figure(figsize=(14, 7))
plt.errorbar(x, y, yerr=yerror, label='Sampled')
plt.plot(x, y_th, label='Analytical curve ')
plt.plot(x_num,(psi[0])**2, label='Numerical curve')
plt.xlabel("x", fontsize=13)
plt.ylabel(r"$| \psi \,(x)|^2$", fontsize=13)
plt.legend(loc=1)
plt.grid(True)
plt.xlim((-2.7,2.7))
plt.show()
```
