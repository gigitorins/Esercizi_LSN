```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import math
import scipy.stats
from itertools import cycle
from IPython.display import clear_output
import time
import scipy
from scipy.optimize import curve_fit
import subprocess
import os

CWD = os.getcwd() # Current work directory
os.chdir("%s" % CWD)

mu = 0.803341  # Optimized parameters from 8.2
sigma = 0.616479

def wave_function_sq(x, mu, sigma):
    a = np.exp(-(x-mu)**2./(2.*sigma**2.)) + np.exp(-(x+mu)**2./(2.*sigma**2.)) 
    b = np.sqrt(2.*np.sqrt(np.pi)*sigma*(1. + np.exp(-(mu/sigma)**2.)))
    return  a/b 

def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar, m, a = 1, 1, 10
N = 1000 # number of iterations
x_num = np.linspace(-a/2, a/2, N)
dx = x_num[1] - x_num[0] # Step size
V = Vpot(x_num)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2
CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V) # Hamiltonian matrix
E,psi = np.linalg.eigh(H) # Compute eigenvectors and their eigenvalues
psi = np.transpose(psi) # Take the transpose & normalize
psi = psi/np.sqrt(dx)
```

Lo scopo di questo esercizio è simulare il sistema utilizzando gli algoritmi Path Integral Ground State (PIGS) (simulando il sistema a $T=0$) e Path Integral Monte Carlo (PIMC) ($T>0$).

PIGS considera la corrispondenza uno a uno tra il percorso che compie una particella quantistica e un polimero aperto classico; ho modificato il codice inserendo il potenziale $V(x)=x^4+5/2x^2$ e le sue derivate prima e seconda. Il sistema viene simulato usando:
1. una funzione d'onda costante $\Psi_T(x)=1$;
2. $\Psi_T^{\sigma,\mu}(x)$ con i parametri ottimizzati dell'*Esercizio 8.2*.

I risultati dell'algoritmo PIGS per entrambe le funzioni d'onda sono confrontati con quelli dell'esercizio precedente. 

## PIGS con funzione d'onda costante

Utilizzando nel codice un tempo di propagazione immaginario $\tau=8.0$ e una funzione d'onda costante la PDF viene riprodotta in modo molto preciso; già riducendo il tempo immaginario a $\tau=2.0$ l'istogramma inizia a discostarsi dai valori numerici.

```python
fnames = ['kinetic.dat', 'potential.dat', 'probability.dat'] # Nomi dei file di output
start = time.time()

subprocess.run("make clean", shell=True, stdout=subprocess.PIPE) # Pulisco i file main.exe, *.o
arguments = str(0)+" "+str(0)+" "+str(sigma)+" "+str(mu)+" "+str(8.0)
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
x1, y1, yerror1 = (np.genfromtxt(fnames[2], usecols=kk, autostrip=True) for kk in range(3))
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE) # Pulisco i file main.exe, *.o
arguments = str(0)+" "+str(0)+" "+str(sigma)+" "+str(mu)+" "+str(2.0)
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
x2, y2, yerror2 = (np.genfromtxt(fnames[2], usecols=kk, autostrip=True) for kk in range(3))
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE) # Pulisco i file main.exe, *.o
arguments = str(0)+" "+str(0)+" "+str(sigma)+" "+str(mu)+" "+str(1.0)
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
x3, y3, yerror3 = (np.genfromtxt(fnames[2], usecols=kk, autostrip=True) for kk in range(3))

end = time.time()
print('-- Computational time: {:.2f} sec '.format(end-start))
```

```python
fig = plt.figure(figsize=(16, 7))
y_th = (wave_function_sq(x1, mu, sigma))**2
plt.plot(x1, y_th, label='VMC')
plt.errorbar(x1, y1, yerr=yerror1, label=r"PIGS, const WF, $\tau=8.0$")
plt.errorbar(x2, y2, yerr=yerror2, label=r"PIGS, const WF, $\tau=2.0$")
plt.errorbar(x3, y3, yerr=yerror3, label=r"PIGS, const WF, $\tau=1.0$")
plt.plot(x_num,(psi[0])**2, label='Valore numerico')
plt.grid(True)
plt.legend(loc=0, fontsize=12)
plt.xlabel('x', fontsize=14)
plt.ylabel('P(x)', fontsize=14)
plt.title('PIGS con funzione d\'onda costante', fontsize=16)
plt.show()
```

## PIGS con funzione d'onda di prova

Con una funzione d'onda non costante $\Psi_T^{\sigma,\mu}(x)$ (più vicina alla $\Psi(x)$ "vera"), il tempo di propagazione immaginario può essere ridotto fino a $\tau=0.5$ pur ottenendo un'approssimazione altrettanto fedele. Diminuendo ulteriormente $\tau$ la $p(x)$ si allontana da quella desiderata.

```python
def replace_line(file_name, line_num, text):
    with open(file_name, 'r') as file:
        data = file.readlines()
    data[line_num] = text
    with open(file_name, 'w') as file:
        file.writelines(data)
        
# ************** tau = 8 ************** # Funzione d'onda costante
subprocess.run("cp input0.pigs input.pigs", shell=True)
replace_line('input.pigs', 4, 'brownianMotionReconstructions\t\t120\n') # moto browniano
replace_line('input.pigs', 5, 'delta_translation\t\t\t0.6\n') # traslazione
replace_line('input.pigs', 6, 'brownianBridgeReconstructions\t\t60\n') # ponte browniano
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE) # Pulisco i file main.exe, *.o
arguments = str(0)+" "+str(0)+" "+str(sigma)+" "+str(mu)+" "+str(8.0)
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
x1, y1, yerror1 = (np.genfromtxt(fnames[2], usecols=kk, autostrip=True) for kk in range(3))

# ************** tau = 2 **************
replace_line('input.pigs', 4, 'brownianMotionReconstructions\t\t220\n') # moto browniano
replace_line('input.pigs', 5, 'delta_translation\t\t\t1.1\n') # traslazione
replace_line('input.pigs', 6, 'brownianBridgeReconstructions\t\t280\n') # ponte browniano
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE) # Pulisco i file main.exe, *.o
arguments = str(0)+" "+str(1)+" "+str(sigma)+" "+str(mu)+" "+str(2.0)
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
x2, y2, yerror2 = (np.genfromtxt(fnames[2], usecols=kk, autostrip=True) for kk in range(3))

# ************** tau = 0.5 **************
replace_line('input.pigs', 5, 'delta_translation\t\t\t2.0\n') # traslazione
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE) # Pulisco i file main.exe, *.o
arguments = str(0)+" "+str(1)+" "+str(sigma)+" "+str(mu)+" "+str(0.5)
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
x3, y3, yerror3 = (np.genfromtxt(fnames[2], usecols=kk, autostrip=True) for kk in range(3))

# ************** tau = 0.4 **************
replace_line('input.pigs', 5, 'delta_translation\t\t\t3.0\n') # traslazione
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE) # Pulisco i file main.exe, *.o
arguments = str(0)+" "+str(1)+" "+str(sigma)+" "+str(mu)+" "+str(0.4)
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
x4, y4, yerror4 = (np.genfromtxt(fnames[2], usecols=kk, autostrip=True) for kk in range(3))
```

```python
fig = plt.figure(figsize=(16, 7))
y_th = (wave_function_sq(x1, mu, sigma))**2
plt.plot(x1, y_th, label='VMC')
plt.errorbar(x1, y1, yerr=yerror1, label=r"PIGS, const WF, $\tau=8.0$")
plt.errorbar(x2, y2, yerr=yerror2, label=r"PIGS, non const WF, $\tau=2.0$")
plt.errorbar(x3, y3, yerr=yerror3, label=r"PIGS, non const WF, $\tau=0.5$")
plt.errorbar(x4, y4, yerr=yerror4, label=r"PIGS, non const WF, $\tau=0.4$")
plt.plot(x_num,(psi[0])**2, label='Valore numerico')
plt.grid(True)
plt.legend(loc=0, fontsize=12)
plt.xlabel('x', fontsize=14)
plt.ylabel('P(x)', fontsize=14)
plt.title('PIGS con funzione d\'onda di prova', fontsize=16)
plt.xlim(-3,3)
plt.show()
```

Nei grafici sottostanti è mostrato l'andamento dei valori medi dell'energia potenziale e di quella cinetica per le due diverse $\Psi(x)$. Possiamo notare che se $\Psi(x)$ è costante nelle prime e ultime time slice il potenziale è molto diverso da quello ottenuto utilizzando $\Psi(x)=\Psi_T^{\sigma,\mu}(x)$. L'energia cinetica invece ha valori compatibili nei due casi.

L'energia totale media è banalmente la somma di energia potenziale e cinetica; la sua incertezza si ricava con la propagazione dell'errore. Sovrapponendo il grafico di $\langle H\rangle$ col valore finale calcolato con il metodo VMC si ha una compatibilità con il valore di VCM per tutte le time slice.
