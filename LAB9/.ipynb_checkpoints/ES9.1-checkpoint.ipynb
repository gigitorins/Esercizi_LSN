```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import matplotlib.mlab as mlab
import numpy as np
import math
import scipy.stats
from itertools import cycle
from IPython.display import clear_output
import time
import scipy
from scipy.optimize import curve_fit
import subprocess
import os
import matplotlib.collections as collections
import matplotlib.ticker as ticker
```

Il problema del commesso viaggiatore è risolvibile con la tecnica degli algoritmi genetici. Rappresentando ogni percorso tramite un vettore monodimensionale contenente l'ordine delle città da visitare, la fitness dell'algoritmo è rappresentata dalla distanza percorsa dal commesso viaggiatore ($\vec x_i$ è la posizione della $i$-esima città)

$$ L_1(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N) = \sum_{i=1}^{N} |\vec{x}_i-\vec{x}_{i+1}| $$

Tale funzione rappresenta il costo che si vuole minimizzare. A partire da una popolazione iniziale formata da un numero $n_\text{pop}$ di individui (pari a $n_\text{pop} = 3000$), si va incontro ad un processo di *selezione* basato sulla fitness; la selezione serve a scegliere gli individui adatti alla riproduzione (*crossover*), che avviene con probabilità $p_\text{c}$. Ogni individuo della nuova popolazione generata può subire delle mutazioni (cambiano l'ordine con cui le diverse città compaiono al suo interno) con probabilità $p_\text{m}$.

I valori di $p_\text{m}$ e $p_\text{c}$ scelti sono: $p_\text{m} = 0.5$ e $p_\text{c} = 0.7$. Il numero di generazioni fissato è $n_\text{gen}=200$.

## Configurazioni iniziali

I due casi affrontati sono:
1) 32 città generate su una circonferenza centrata in $x = r = 5$ e di raggio $r = 5$.
2) 32 città generate all'interno di un quadrato di lato $l = 10$. 

Nei grafici sono mostrate le posizioni (generate casualmente) delle 32 città per entrambe le configurazioni.

```python
fig = plt.figure(figsize=(16, 8))
rect = plt.Rectangle((0., 0.), 10, 10, facecolor='none', edgecolor='b',linewidth=2.0, alpha=1)
circ = plt.Circle((5., 5.), 5, facecolor='none', edgecolor='b',linewidth=2.0, alpha=1)
patches = [rect,circ]
filenames = ["config_s.dat", "config_c.dat"]
gnames = ["quadrata", "circolare"]
x1, x2, y1, y2 = ([] for i in range(4))
xtot, ytot  = [x1, x2], [y1, y2]

for j,file in enumerate(filenames):
    ax = fig.add_subplot(1, 2, j+1)
    xtot[j], ytot[j] = (np.genfromtxt(file, usecols=kk, autostrip=True) for kk in range(2))
    for i in range(len(xtot[j])):
        plt.plot(xtot[j][i], ytot[j][i],'ro')
        plt.annotate(i+1,(xtot[j][i],ytot[j][i]),xytext=(xtot[j][i]+0.05, ytot[j][i]+0.05))
    ax.add_patch(patches[j])
    ax.grid(True)  
    ax.set_xlabel("X", fontsize=15)
    ax.set_ylabel("Y", fontsize=15)
    ax.set_xlim((-1,11))
    ax.set_ylim((-1,11))
    ax.set_title("Configurazione %s" % gnames[j], fontsize=16)
    
fig.tight_layout()
plt.show()
```

## GA per configurazione quadrata

```python
start = time.time()
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
arguments = "0"
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
end = time.time()
print('-- Computational time: {:.2f} sec '.format(end-start))
ordS = np.genfromtxt("solution.dat")
walksS = np.genfromtxt("walks.dat")
avgsS = np.genfromtxt("averages.dat")
```

```python
fig = plt.figure(figsize=(10, 10))
rect = plt.Rectangle((0., 0.), 10, 10, facecolor='none', edgecolor='b',linewidth=2.0, alpha=1)
width = 0.001
h_w = 150*width
h_l = h_w
x, y = xtot[0], ytot[0]

for i in range(len(x)):
    ap = int(ordS[i]-1)
    plt.plot(x[ap], y[ap],'ro')
    plt.annotate(int(ordS[i]),(x[ap],y[ap]),xytext=(x[ap]+0.01, y[ap]+0.13), size=13)
    if i == len(x)-1:
        plt.arrow(x[ap], y[ap], x[0]-x[ap], y[0]-y[ap], width=width, head_width=h_w, head_length=h_l,
                 length_includes_head=True, alpha=0.5)
    else:
        ap2 = int(ordS[i+1]-1)
        plt.arrow(x[ap], y[ap], x[ap2]-x[ap], y[ap2]-y[ap], width=width, head_width=h_w, head_length=h_l,
                  length_includes_head=True, alpha=0.5)
        
ax = plt.gca()
ax.add_patch(rect)
plt.plot(x[0],y[0],'*', markersize=16, color='orange')
plt.grid(True)  
plt.xlabel("X", fontsize=15)
plt.ylabel("Y", fontsize=15)
plt.xlim((-1,11))
plt.ylim((-1,11))
plt.title("Best walk", fontsize=17)
plt.show()
```

Il grafico in alto rappresente il miglior percorso trovato; il grafico sottostante mostra l'andamento della funzione costo $L_1$ del miglior individuo della popolazione rispetto all'andamento della media di $L_1$ sulla migliore metà della popolazione, in funzione della generazione. Dal secondo grafico si osserva che la fitness ha miglioramenti trascurabili dalla generazione 80 in poi.

```python
fig = plt.figure(figsize=(14, 8))
plt.plot(walksS, label="Best walk")
plt.plot(avgsS, label='Best walk medio sulla migliore metà della popolazione')
print("Best walk: ", min(walksS))
plt.xlabel("Generazione",fontsize=13)
plt.ylabel("Distanza",fontsize=13)
plt.legend(loc='best', prop={'size': 12})
plt.grid(True)
plt.show()
```

## GA per configurazione circolare

```python
start = time.time()
subprocess.run("make clean", shell=True, stdout=subprocess.PIPE)
arguments = "1"
subprocess.run('make esegui ARGS="%s"' % arguments, shell=True)
end = time.time()
print('-- Computational time: {:.2f} sec '.format(end-start))
ordC = np.genfromtxt("solution.dat")
walksC = np.genfromtxt("walks.dat")
avgsC = np.genfromtxt("averages.dat")
```

```python
fig = plt.figure(figsize=(12, 12))
circ = plt.Circle((5., 5.), 5, facecolor='none', edgecolor='b',linewidth=2.0, alpha=1)
width = 0.001
h_w = 150*width
h_l = h_w
x, y = xtot[1], ytot[1]


for i in range(len(x)):
    ap = int(ordC[i]-1)
    plt.plot(x[ap], y[ap],'ro')
    plt.annotate(int(ordC[i]),(x[ap],y[ap]),xytext=(x[ap]+0.01, y[ap]+0.13), size=13)
    if i == len(x)-1:
        plt.arrow(x[ap], y[ap], x[0]-x[ap], y[0]-y[ap], width=width, head_width=h_w, head_length=h_l,
                 length_includes_head=True, alpha=0.5)
    else:
        ap2 = int(ordC[i+1]-1)
        plt.arrow(x[ap], y[ap], x[ap2]-x[ap], y[ap2]-y[ap], width=width, head_width=h_w, head_length=h_l,
                  length_includes_head=True, alpha=0.5)
        
ax = plt.gca()
ax.add_patch(circ)
plt.plot(x[0],y[0],'*', markersize=16, color='orange')
plt.grid(True)  
plt.xlabel("X", fontsize=15)
plt.ylabel("Y", fontsize=15)
plt.xlim((-1,11))
plt.ylim((-1,11))
plt.title("Best walk", fontsize=17)
plt.show()
```

Il grafico in alto rappresente il miglior percorso trovato; il grafico sottostante mostra l'andamento della funzione costo $L_1$ del miglior individuo della popolazione rispetto all'andamento della media di $L_1$ sulla migliore metà della popolazione, in funzione della generazione. Il commesso viaggiatore compie effettivamente il percorso più breve, ossia percorre la circonferenza per minimizzare il tragitto. Dal secondo grafico si osserva che la fitness ha miglioramenti trascurabili dalla generazione 50 in poi.

```python
fig = plt.figure(figsize=(14, 8))
plt.plot(walksC, label="Best walk")
plt.plot(avgsC, label='Best walk medio sulla migliore metà della popolazione')
print("Misura della circonferenza: ", 2*np.pi*5.)
plt.axhline(2*np.pi*5., color='k',label="Misura della circonferenza", alpha = 0.6)
print("Best walk: ", min(walksC))
plt.xlabel("Generazione",fontsize=13)
plt.ylabel("Distanza",fontsize=13)
plt.legend(loc='best', prop={'size': 12})
plt.grid(True)
plt.show()
```
